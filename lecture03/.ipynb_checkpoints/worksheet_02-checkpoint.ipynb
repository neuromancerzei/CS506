### Effective Programming

a) What is a drawback of the top down approach?

The fully functioning system will only emerge after most, if not all, components are implemented. And you can't test the code along the way: you don't have a runnable code until the end. This is in contrast to a bottom-up approach, where components can often be tested and used as soon as they are developed.

b) What is a drawback of the bottom up approach?

```python
class Board:
    def __init__(self):
        self.board = [['_' for i in range(8)]for j in range(8)]
        self.solutions = []
    
    def __repr__(self):
        res = ""
        for row in self.board:
            for obj in row:
                res+= obj + " "
            res += "\n"
        return res
    
    def is_legal(self, row, col):
        # Check for a queen in the same row or column
        for i in range(8):
            if self.board[row][i]=="Q" or self.board[i][col]=="Q":
                return False
        # Check for a queen on the diagonals
        for i in range(8):
            if (row + i < 8 and col + i < 8 and self.board[row + i][col + i]=="Q") or \
               (row - i >= 0 and col - i >= 0 and self.board[row - i][col - i]=="Q") or \
               (row + i < 8 and col - i >= 0 and self.board[row + i][col - i]=="Q") or \
               (row - i >= 0 and col + i < 8 and self.board[row - i][col + i]=="Q"):
                return False
        return True
    
    def set_queen(self,row,col):
        self.board[row][col]="Q"
    
    def unset_queen(self,row):
        self.board[row]=["_" for i in range(8)]
    
    def solve_util(self, current_row):
        # Base case: If all queens are placed
        if current_row == 8:
            print("Find a solution:")
            print(self)
            return
        # Try placing queen in all columns one by one
        for i in range(8):
            if self.is_legal(current_row, i):
                self.set_queen(current_row, i)
                self.solve_util(current_row + 1)
                self.unset_queen(current_row)  # backtrack
    def solve(self):
        self.solve_util(0)


testBoard = Board()
#print(testBoard)
testBoard.solve()
```
